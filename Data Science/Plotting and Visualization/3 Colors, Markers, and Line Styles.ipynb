```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Colors, Markers, and Line Styles #to plot x versus y with green dashes
fig=plt.figure()
ax=fig.add_subplot()
# ax.plot(np.random.standard_normal(30).cumsum(), color="black", linestyle="dashed", marker='o')
data=np.random.standard_normal(30).cumsum()
ax.plot(data,color="black",linestyle="dashed", label="Default") #labeled the plot
ax.plot(data,color="black",linestyle="dashed", label="steps-post", drawstyle="steps-post")
ax.legend()
```
