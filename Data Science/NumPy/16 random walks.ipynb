```python
#Random Walks
import matplotlib.pyplot as plt
import random
import numpy as np
position=0
walk=[position]
nsteps=1000
for _ in range(nsteps):
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)
plt.plot(walk[:1000])
rng=np.random.default_rng(seed=12345)
draws=rng.integers(0,2,size=nsteps)
steps=np.where(draws==0,1,-1)
walk=steps.cumsum()
walk.min(),walk.max() #min and max point
(np.abs(walk)>=10).argmax() #finding where the distance is 10 occure first time

```
